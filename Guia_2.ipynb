### **Caso de Negocio: Análisis de Deserción de Clientes en una Empresa de Telecomunicaciones**

**Contexto**

En la industria de las telecomunicaciones, la retención de clientes es un factor crítico para la sostenibilidad del negocio. La competencia es feroz y adquirir nuevos clientes suele ser más costoso que mantener a los actuales. En este análisis, exploraremos un conjunto de datos que contiene información detallada sobre clientes que han abandonado el servicio (churn) y aquellos que permanecen activos.

Nuestro objetivo es identificar los factores clave que influyen en la deserción, comprender patrones de comportamiento y generar estrategias efectivas para reducir la tasa de abandono.

La empresa ha experimentado un aumento en la tasa de deserción de clientes, lo que ha provocado:
- Pérdida de ingresos recurrentes.
- Incremento en los costos de adquisición de nuevos clientes.
- Menor estabilidad en la base de clientes a largo plazo.

Para abordar esta problemática, es crucial identificar las razones detrás de la deserción y desarrollar estrategias para mejorar la retención de clientes.



**Descripción de las variables del dataset telecom_churn**

El dataset telecom_churn contiene información detallada sobre clientes de una empresa de telecomunicaciones, incluyendo datos generales, planes contratados, uso del servicio telefónico y llamadas al servicio al cliente. Su propósito principal es analizar patrones de deserción de clientes, identificados a través de la variable churn, que indica si un cliente ha abandonado la empresa (1) o sigue siendo cliente (0).

Dentro del dataset, encontramos información general como el estado (state) donde reside el cliente, el código de área (area code), y el número de teléfono (phone number), aunque esta última variable no aporta información útil para el análisis, ya que es un identificador único.

Además, el dataset registra el tiempo que un cliente ha estado en la empresa a través de la variable account length, lo que puede ayudar a analizar si la duración del contrato influye en la deserción. También se incluyen detalles sobre los planes contratados, como si el cliente tiene un plan internacional (international plan), que le permite realizar llamadas internacionales, o un buzón de voz (voice mail plan), que le permite recibir mensajes de voz.

En cuanto al uso del servicio, se registran datos detallados sobre el tiempo en llamadas y los costos asociados. Se divide en tres períodos del día: diurno (total day minutes, total day calls, total day charge), vespertino (total eve minutes, total eve calls, total eve charge) y nocturno (total night minutes, total night calls, total night charge), lo que permite evaluar si hay patrones de consumo que influyen en la deserción. También se incluye información sobre el uso del servicio internacional, con variables como total intl minutes (minutos en llamadas internacionales), total intl calls (cantidad de llamadas internacionales) y total intl charge (costos por llamadas internacionales).

Otro aspecto clave del dataset es el número de llamadas al servicio al cliente (customer service calls), ya que una mayor cantidad de llamadas puede indicar insatisfacción y estar relacionada con la decisión del cliente de abandonar la empresa.

Finalmente, la variable más importante del análisis es churn, que indica si un cliente ha desertado de la empresa. A partir de esta variable, podemos analizar qué factores influyen en la deserción y encontrar patrones en los clientes que tienen mayor probabilidad de abandonar el servicio.

**Análisis de Churn y Factores Relacionados**

Calcula el porcentaje de clientes que han desertado (churn = 1).

Identifica si los clientes con plan internacional (international plan) tienen mayor tasa de deserción.

Identifica si los clientes con buzón de voz (voice mail plan) tienen menor tasa de deserción.

```python
# Solución propuesta

print('\n Porcentaje de clientes que han desertado')
df_porcentaje_desercion = df['churn'].value_counts(normalize=True) * 100 #Porcentaje de clientes que han desertado, siendo True el porcentaje de desercion
print(df_porcentaje_desercion)


print('\n Tasa de desercion de los clientes con plan internacional')
df_international_plan = df[df['international plan'] == 'yes'] #Filtrado clientes con plan internacinal
df_international_plan_desercion = df_international_plan['churn'].value_counts(normalize=True) * 100 #Porcentaje de clientes con plan internacional que han desertado
print(df_international_plan_desercion)
print('Los clientes con plan internacional tienen una tasa de desercion mayor')

print('\n Tasa de desercion de los clientes con buzon de voz')
df_voice_mail_plan = df[df['voice mail plan'] == 'yes'] #Filtrado de clientes con buzon de voz
df_voice_mail_plan_desercion = df_voice_mail_plan['churn'].value_counts(normalize=True) * 100 #Taza de desercion de clientes con buzón de voz
print(df_voice_mail_plan_desercion)
print('Los clientes con buzon de voz tienen menor tasa de desercion')
```

**Análisis de la Duración del Servicio y Deserción**

¿Cuál es la duración promedio de la cuenta (account length) entre clientes que desertaron y los que permanecen?

¿Los clientes con cuentas más antiguas tienen más probabilidades de desertar?

```python
# Solución propuesta

print('\n Duracion promedio de la cuenta de los clientes que desertaron')
df_desercion = df[df['churn'] == 1] #Filtrado de clientes que desertaron
print(df_desercion['account length'].mean()) #Duracion promedio de la cuenta de los clientes que desertaron

print('\n Duracion promedio de la cuenta de los clientes que no desertaron')
df_no_desercion = df[df['churn'] == 0] #Filtrado de clientes que no desertaron
print(df_no_desercion['account length'].mean())
```

Rta: Con este analisis odemos observar que los clientes con cuentas mas antiguas tienden a tener mas probabilidades de desertar.

**Relación entre Deserción y Uso del Servicio**

Compara la cantidad de minutos usados en llamadas diurnas (total day minutes) entre clientes con y sin churn.

Compara la cantidad de minutos usados en llamadas nocturnas (total night minutes).

Compara el número total de llamadas (total day calls) entre clientes con y sin churn.

```python
# Solución propuesta

print('\n Cantidad de minutos usados en llamadas diurnas entre clientes con churn')
df_llamadas_diurnas_con = df[df['churn'] == 1] #Filtrado de clientes con churn
print(df_llamadas_diurnas_con['total day minutes'].mean()) #Cantidad de minutos usados en llamadas diurnas de clientes que desertaron

print('Cantidad de minutos usados en llamadas diurnas entre clientes sin churn')
df_llamadas_diurnas_sin = df[df['churn'] == 0] #Filtrado de clientes sin churn
print(df_llamadas_diurnas_sin['total day minutes'].mean())
print('Los clientes que desertaron usaron mas minutos en llamadas diurnas')


print('\n Cantidad de minutos usados en llamadas nocturnas entre clientes con churn')
df_llamadas_nocturnas_con = df[df['churn'] == 1] #Filtrado de clientes con churn
print(df_llamadas_nocturnas_con['total night minutes'].mean()) #Cantidad de minutos usados en llamadas nocturnas de clientes que desertaron

print('Cantidad de minutos usados en llamadas nocturnas entre clientes sin churn')
df_llamadas_nocturnas_sin = df[df['churn'] == 0] #Filtrado de clientes sin churn
print(df_llamadas_nocturnas_sin['total night minutes'].mean())
print('Los clientes que desertaron usaron mas minutos en llamadas nocturnas')

print('\n Cantidad de llamadas totales entre clientes con churn')
df_llamadas_totales = df[df['churn'] == 1] #Filtrado de clientes con churn
print(df_llamadas_totales['total day calls'].mean()) #Cantidad de llamadas totales de clientes que desertaron

print('Cantidad de llamadas totales entre clientes sin churn')
df_llamadas_totales_sin = df[df['churn'] == 0] #Filtrado de clientes sin churn
print(df_llamadas_totales_sin['total day calls'].mean())
print('Los clientes que desertaron realizaron mas llamadas')
```

**Impacto de las Llamadas al Servicio al Cliente en la Deserción**

Calcula el número promedio de llamadas al servicio al cliente (customer service calls) entre clientes que desertaron y los que no.

Divide los clientes en dos grupos:

- Grupo 1: Clientes que llamaron más de 3 veces al servicio al cliente.
- Grupo 2: Clientes que llamaron 3 veces o menos.
Compara la tasa de churn entre ambos grupos.

```python
# Solución propuesta

print('\n Cantidad de llamadas al servicio al cliente entre clientes que desertaron')
df_llamadas_cliente_con = df[df['churn'] == 1] #Filtrado de clientes con churn
print(df_llamadas_cliente_con['customer service calls'].mean()) #Cantidad de llamadas al servicio al cliente de clientes que desertaron

print('Cantidad de llamadas al servicio al cliente entre clientes sin churn')
df_llamadas_cliente_sin = df[df['churn'] == 0] #Filtrado de clientes sin churn
print(df_llamadas_cliente_sin['customer service calls'].mean())

print('\n Grupo 1: Clientes que llamaron mas de 3 veces al servicio al cliente')
df_llamadas_cliente_sup_3 = df[df['customer service calls'] > 3] #Filtrado de clientes que llamaron mas de 3 veces al servicio
print(df_llamadas_cliente_sup_3)
print(df_llamadas_cliente_sup_3['churn'].value_counts(normalize=True) * 100)

print('\n Grupo 2: Clientes que llamaron 3 veces o menos al servicio al cliente')
df_llamadas_cliente_inf_3 = df[df['customer service calls'] <= 3] #Filtrado de clientes que llamaron 3 veces o menos al servicio
print(df_llamadas_cliente_inf_3)
print(df_llamadas_cliente_inf_3['churn'].value_counts(normalize=True) * 100)
```

**Análisis del Costo de las Llamadas y Churn**

Compara el costo total de llamadas diurnas (total day charge) entre clientes con y sin churn.

Compara el costo total de llamadas nocturnas (total night charge).

¿Los clientes con mayor gasto en llamadas internacionales (total intl charge) tienen más probabilidades de desertar?

```python
# Solución propuesta

print('\n Costo total de llamadas diurnas entre clientes con churn')
df_costo_llamadas_diurnas_con = df[df['churn'] == 1] #Filtrado de clientes con churn
print(df_costo_llamadas_diurnas_con['total day charge'].mean())

print('Costo total de llamadas diurnas entre clientes sin churn')
df_costo_llamadas_diurnas_sin = df[df['churn'] == 0] #Filtrado de clientes sin churn
print(df_costo_llamadas_diurnas_sin['total day charge'].mean())
print('Los clientes que desertaron gastaron mas dinero en llamadas diurnas')

print('\n Costo total de llamadas nocturnas entre clientes con churn')
print(df_costo_llamadas_diurnas_con['total night charge'].mean())

print('Costo total de llamadas nocturnas entre clientes sin churn')
print(df_costo_llamadas_diurnas_sin['total night charge'].mean())
print('Los clientes que desertaron gastaron mas dinero en llamadas nocturnas')

print('\n Costo total de llamadas internacionales entre clientes con churn')
print(df_costo_llamadas_diurnas_con['total intl charge'].mean())

print('Costo total de llamadas internacionales entre clientes sin churn')
print(df_costo_llamadas_diurnas_sin['total intl charge'].mean())
print('Podemos observar que Los clientes con mayor gasto en llamadas internacionales desertaron')
```
